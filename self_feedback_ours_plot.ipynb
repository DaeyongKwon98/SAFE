```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.lines import Line2D

# 1. 데이터 설정
data = {
    'error_type': ['Redundancy', 'Off-topic', 'Logical Fallacy', 'Unsupported', 'Overthinking', 
                  'Information Miss', 'Contradictory', 'Wrong Conclusion', 'Premature Attribution', 'Inefficiency'],
    'avg_ratio': [0.081396, 0.077267, 0.039256, 0.030169, 0.022789, 0.012997, 0.009758, 0.007899, 0.002482, 0.001442]
}

df = pd.DataFrame(data)

# --- [중요] 전체 에러 합계를 100%로 재정규화 ---
total_error_sum = df['avg_ratio'].sum()
df['pct'] = (df['avg_ratio'] / total_error_sum) * 100 

# 2. Taxonomy 카테고리 매핑
category_map = {
    'Redundancy': 'Procedural', 'Off-topic': 'Procedural', 'Overthinking': 'Procedural', 'Inefficiency': 'Procedural',
    'Unsupported': 'Attribution', 'Premature Attribution': 'Attribution', 'Information Miss': 'Attribution', 'Contradictory': 'Attribution',
    'Logical Fallacy': 'Logical', 'Wrong Conclusion': 'Final Answer'
}
df['category'] = df['error_type'].map(category_map)

# 카테고리 순서 정의 (위에서부터 Procedural -> Final Answer 순)
category_order = ['Final Answer', 'Logical', 'Attribution', 'Procedural']
df['category'] = pd.Categorical(df['category'], categories=category_order, ordered=True)
df = df.sort_values(by=['category', 'pct'], ascending=[True, True])

# 3. 시각화 설정
colors = {'Procedural': '#3498db', 'Attribution': '#e67e22', 'Logical': '#2ecc71', 'Final Answer': '#e74c3c'}
fig, ax = plt.subplots(figsize=(10, 7))
bars = ax.barh(df['error_type'], df['pct'], color=[colors[c] for c in df['category']])

# 폰트 크기 대폭 상향
ax.tick_params(axis='y', labelsize=20)
ax.tick_params(axis='x', labelsize=16)
ax.set_xlabel('Proportion among Errors (%)', fontsize=18, fontweight='medium', labelpad=10)
# ax.set_title('Distribution of Reasoning Errors', fontsize=16, fontweight='bold', pad=20)

# 4. 텍스트 배치 (범위 밖으로 나가지 않게 넉넉히 설정)
max_pct = df['pct'].max()
ax.set_xlim(0, max_pct * 1.2) 

for bar in bars:
    width = bar.get_width()
    ax.text(width + (max_pct * 0.02), bar.get_y() + bar.get_height()/2, 
            f'{width:.1f}%', va='center', fontweight='medium', fontsize=16)

# 5. 범례 (카테고리별 합산 100% 확인용)
category_totals = df.groupby('category', observed=True)['pct'].sum().to_dict()
legend_order = ['Procedural', 'Attribution', 'Logical', 'Final Answer']
legend_elements = [
    Line2D([0], [0], color=colors[k], lw=4, label=f"{k} ({category_totals[k]:.1f}%)") 
    for k in legend_order
]
ax.legend(handles=legend_elements, loc='lower right', fontsize=15)

ax.grid(axis='x', linestyle='--', alpha=0.4)
ax.set_axisbelow(True)
plt.tight_layout()
plt.savefig("reasoning_error_distribution.png", dpi=300, bbox_inches='tight', pad_inches=0.01)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.lines import Line2D

# 1. 데이터 설정
data = {
    'error_type': [
        'Redundancy', 'Logical Fallacy', 'Information Miss', 'Off-topic', 
        'Unsupported', 'Overthinking', 'Contradictory', 'Wrong Conclusion', 
        'Inefficiency', 'Premature Attribution'
    ],
    'spearman_rho': [
        -0.319682, -0.280533, -0.259545, -0.252253, 
        -0.250892, -0.177399, -0.160511, -0.124600, 
        -0.091987, 0.024929
    ]
}

df = pd.DataFrame(data).sort_values(by='spearman_rho', ascending=True)

# 2. 카테고리 매핑 및 색상
category_map = {
    'Redundancy': 'Procedural', 'Logical Fallacy': 'Logical', 'Information Miss': 'Attribution',
    'Off-topic': 'Procedural', 'Unsupported': 'Attribution', 'Overthinking': 'Procedural',
    'Contradictory': 'Attribution', 'Wrong Conclusion': 'Final Answer',
    'Inefficiency': 'Procedural', 'Premature Attribution': 'Attribution'
}
colors = {'Procedural': '#3498db', 'Attribution': '#e67e22', 'Logical': '#2ecc71', 'Final Answer': '#e74c3c'}

# 3. 시각화
fig, ax = plt.subplots(figsize=(11, 7))
bars = ax.barh(df['error_type'], df['spearman_rho'], 
               color=[colors[category_map[et]] for et in df['error_type']])

# x축 범위: 왼쪽 텍스트 공간 확보 (-0.36까지 확장) 및 오른쪽 여백
ax.set_xlim(-0.36, 0.06) 

ax.axvline(0, color='black', linewidth=1)
ax.set_xlabel("Spearman's Correlation ($\\rho$)", fontsize=15, fontweight='bold')
ax.set_title("Correlation between Error Types and Accuracy", fontsize=15, fontweight='bold', pad=20)
ax.tick_params(axis='y', labelsize=12)

# 값 표시 (텍스트 잘림 방지 최적화)
for bar in bars:
    width = bar.get_width()
    offset = 0.008 
    ax.text(width - offset if width < 0 else width + offset, 
            bar.get_y() + bar.get_height()/2, 
            f'{width:.2f}', va='center', 
            ha='right' if width < 0 else 'left', 
            fontweight='medium', fontsize=14)

# --- [수정 포인트] 좌측 상단 범례 추가 ---
legend_order = ['Procedural', 'Attribution', 'Logical', 'Final Answer']
legend_elements = [Line2D([0], [0], color=colors[k], lw=6, label=k) for k in legend_order]

ax.legend(handles=legend_elements, 
          loc='upper left',          # 좌측 상단 배치
          fontsize=14, 
          frameon=True,              # 배경 박스 활성화
          shadow=False, 
          edgecolor='gray')

ax.grid(axis='x', linestyle='--', alpha=0.4)
ax.set_axisbelow(True)

plt.tight_layout()
plt.savefig("error_correlation_analysis.png", dpi=300, bbox_inches='tight', pad_inches=0.01)
plt.show()
```
